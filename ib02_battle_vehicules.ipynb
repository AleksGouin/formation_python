## Exercice 2

```python
categorie ={
1: "Véhicule en stationnement",
2: "Arbre",
3: "Glissière métallique",
4: "Glissière béton",
5: "Autre glissière",
6: "Bâtiment, mur, pile de pont",
7: "Support de signalisation verticale ou poste d’appel d’urgence",
8: "Poteau",
9: "Mobilier urbain ",
8: "poteau",
10: "Parapet",
11: "Ilot, refuge, borne haute",
12: "Bordure de trottoir",
13: "Fossé, talus, paroi rocheuse",
14: "Autre obstacle fixe sur chaussée",
15: "Autre obstacle fixe sur trottoir ou accotement",
16: "Sortie de chaussée sans obstacle "
    
}

df['obs_enc'] = df.obs.map(categorie)

obs = df.obs_enc.value_counts()
obs.plot.barh(title='Type d\'obstacle fixe')
fig = plt.gcf()
fig.set_size_inches(14, 10)
```

## exo3 : créer bokeh d'un scatter plot

```python
tab = df.pivot_table(columns='obs_enc', index='occutc', values='Num_Acc', aggfunc='count')


sns.heatmap(tab.div(tab.sum().sum()), robust=True, annot=True, cmap='Reds', fmt='.2%')
fig = plt.gcf()
fig.suptitle('Obstacles fixes VS occupants', horizontalalignment='right')
fig.set_size_inches(14, 10)
```
