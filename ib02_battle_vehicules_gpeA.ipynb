```python
df = pd.read_csv(path, encoding="latin1")
```

```python
new_columns = ['Accident_ID', 'Sens_Circulation', 'Categorie_Vehicule', 'Nb_Personnes', 'Obstacle_Fix_Heurte', 'Obstacle_Mobile_Heurte', 'Point_Choc', 'Manoeuvre_Avant', 'Vehicule_ID']
```

```python
list(zip(new_columns, df.columns))
```

```python
df.columns = new_columns
```

```python
dico={
    1 :"Véhicule en stationnement",
    2 :"Arbre",
    3 :"Glissière métallique",
    4 :"Glissière béton",
    5 :"Autre glissière",
    6 :"Bâtiment, mur, pile de pont",
    7 :"Support de signalisation verticale ou poste d’appel d’urgence",
    8 :"Poteau",
    9 :"Mobilier urbain",
    10 :"Parapet",
    11 :"Ilot, refuge, borne haute",
    12 : "Bordure de trottoir",
    13 : "Fossé, talus, paroi rocheuse",
    14 : "Autre obstacle fixe sur chaussée",
    15 : "Autre obstacle fixe sur trottoir ou accotement",
    16 : "Sortie de chaussée sans obstacle"
}
```

```python
df['Obstacle_Fix_Heurte_enc']= df.Obstacle_Fix_Heurte.map(dico)
df.Obstacle_Fix_Heurte_enc.value_counts()

```
