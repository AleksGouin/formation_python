# ModÃ©lisation

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
col_y = "Num_Acc"
```

```python
caracteristiques.shape
```

```python
sans_na = caracteristiques.select_dtypes(exclude="object").drop(["lat", "long"], axis="columns").dropna()#.shape#isna().sum()
```

```python
X = sans_na.drop(col_y, axis="columns")
Y = sans_na[col_y]
```

```python
X_train, X_test      , y_train, y_test       = train_test_split(X      ,Y      , test_size=0.15)
X_train, X_validation, y_train, y_validation = train_test_split(X_train,y_train, test_size=0.03)
X_train.shape, X_validation.shape, y_train.shape, y_validation.shape, X_test.shape, y_test.shape
```

```python
r = []
for i in range(50):
    hyperparametres = {"n_estimators" : 20 , 
                       "n_jobs"       : -1 }
    algo = RandomForestRegressor(**hyperparametres)
    modele = algo.fit(X_train.select_dtypes(exclude=["object"]), y_train)
    score = modele.score(X_test, y_test)
    r.append(score)
    print(score, end=" / ")
```

```python
%matplotlib inline
pd.DataFrame(r).plot(ylim=[0,1])
```
